```python
def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df['Sentiment'] = df['Text'].apply(analyze_sentiment)
```

```python
sentiment_counts = df['Sentiment'].value_counts()
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('vader_lexicon')
nltk.download('punkt')
```

```python
def preprocess_text(text):
    # Tokenize and remove stop words
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    filtered_words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
def analyze_emotion(text):
    sid = SentimentIntensityAnalyzer()
    sentiment_scores = sid.polarity_scores(text)

    # Determine emotion based on sentiment score
    if sentiment_scores['compound'] >= 0.05:
        emotion = 'Positive'
    elif sentiment_scores['compound'] <= -0.05:
        emotion = 'Negative'
    else:
        emotion = 'Neutral'

    return emotion
```

```python
text_data = amazon['description']
```

```python
emotion_review=[]
for text in amazon['description']:
    preprocessed_text = preprocess_text(text)
    emotion = analyze_emotion(preprocessed_text)
    emotion_review.append(emotion)
```

```python
emotion_review
```

```python
amazon['emotion_mining']=emotion_review
```
